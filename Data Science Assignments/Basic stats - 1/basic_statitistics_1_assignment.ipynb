# Standardization of Numerical Variables


* Standardization involves transforming raw data values to a standard scale with a mean of 0 and a standard deviation of 1.
*  This is achieved using the formula:
z=x-mu/sigma.
* Standardization helps reduce the effect of extreme values, improves model performance, and makes data more comparable across different datasets.

* It is commonly used in machine learning, finance, and scientific research to preprocess data before analysis or modeling.

```python
df #Before making z normalization
```

```python
df.describe().columns #Get numerical columns
```

```python
for x in df.describe().columns:
    l = [] #Initilizing the list to append all values to this list later update values using this list
    for y in df[x]:
        mean = df[x].mean() #Calculating the mean of the column
        std = df[x].std() #Calculating the standard deviation of the column
        z = (y-mean)/std #calculating the z value
        l.append(round(z,2)) #append calculated z values to list
    df[x].update(l) #Updating the column with list
df #Displaying the column after normalization
```

# Conversion of Categorical Data into Dummy Variables

* One-hot encoding involves converting categorical data into dummy variables.
* This transformation is necessary because algorithms such as linear regression can only handle continuous or numerical data.
* Categorical data cannot be directly processed by these models.
* One-hot encoding transforms categorical data into a binary format, where each category is represented as a separate binary column.
* This allows algorithms to differentiate between distinct categories and enables accurate modeling and prediction of categorical relationships, thereby improving the overall performance of machine learning models.
